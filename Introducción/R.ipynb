```r
# operaciones matemáticas
print("suma")
print(1 + 2)
print("resta")
print(1 - 2)
print("multiplicación")
print(3 * 2)
print("división")
print(8 / 2)
```

```r
# operaciones matemáticas con variables
# definimos las variables
x <- 1
y <- 3

# operamos
print("suma")
print(x + y)
print("resta")
print(x - y)
print("multiplicación")
print(x * y)
print("división")
print(y / x)
```

```r
# condicionales
x <- 3

if (x < 5) {
    # imprimimos que x es menor que 5 si x es menor que 5
    print("x es menor que 5")
}
```

```r
# Bucles
ciclos <- 10
x <- 0

# hacer 10 repeticiones
for (i in 1:ciclos) {
    # sumar 1 a x en cada ciclo
    x <- x + 1

    # imprimir el valor de x
    print(stringr::str_interp("El valor de x es ${x}"))
}
```

## Funciones en R

```r
# Definimos una función (power, elevar un número a cualquier exponente)
pwr <- function(base, exponente) {
    return(base^exponente)
}

# Probamos la función
pwr(2, 5)
```
